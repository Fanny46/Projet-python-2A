```python
#évolution de la taille des fichiers dvf depuis 2018

années = ['2018', '2019', '2020', '2021', '2022', '2023']
taille_fichier_dvf = []

for année in années:
    dvf = pd.read_csv(f'Données_brutes/DVF_75_{année}.csv', low_memory=False)
    taille = dvf.shape[0]
    taille_fichier_dvf.append(taille)

plt.plot(années, taille_fichier_dvf)
plt.title("Evolution de la taille du fichier dvf depuis 2018")
plt.xlabel("Année")
plt.ylabel("Taille fichier dvf")
#plt.savefig("Evolution taille dvf 2018-2023.png", dpi=300"""
```

En fait, le fichier 2018 ne commence qu'en Juillet, c'est pour ça qu'il contient moins de ventes.

```python
#évolution du nombre de dépendances dans les fichiers dvf depuis 2018

années = ['2018', '2019', '2020', '2021', '2022', '2023']
nb_dep_fichier_dvf = []

for année in années:
    dvf = pd.read_csv(f'Données_brutes/DVF_75_{année}.csv', low_memory=False)
    dvf = dvf.loc[dvf['type_local']=='Dépendance']
    taille = dvf.shape[0]
    nb_dep_fichier_dvf.append(taille)

plt.plot(années, nb_dep_fichier_dvf)
plt.title("Evolution du nombre de dépendances dans les fichier dvf depuis 2018")
plt.xlabel("Année")
plt.ylabel("Nombre de dépendances")
```
