```python
"""On cherche les 100 meilleurs lycées seulement : on prend la liste du classement figaro"""

import requests
from bs4 import BeautifulSoup

# URL du classement des lycées à Paris
url = 'https://classement-lycees.etudiant.lefigaro.fr/ville-paris+75056/'

# Récupérer le contenu HTML de la page
response = requests.get(url)
html_content = response.content

# Utiliser BeautifulSoup pour parser le HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Trouver les noms des lycées dans le classement
lycees_classement = [lycee.find('b').text.strip() for lycee in soup.find_all('td', class_='school')]

# Passer tous les noms en majuscule, retirer la partie 'LYCEE ' des noms
lycees_liste = [unidecode(lycee.replace('Lycée ', '').replace('LYCEE ', '').replace('-', '').replace(' ', '').upper()) for lycee in lycees_classement]

#changer l'écriture de Henri IV
for i in range(len(lycees_liste)):
    lycees_liste[i] = lycees_liste[i].replace('HENRI-IV', 'HENRI IV')

#afficher la liste
#lycees_liste

"""
La liste a été enregistrée ici : 
3.0) Enrichissement données/Données/liste_meilleurs_lycées.pkl
"""
```

```python
"""Charger la liste des meilleurs lycées"""

with open(emplacement_liste, 'rb') as fichier:
    liste_meilleurs_lycées = pickle.load(fichier)[:40] #prendre les x meilleurs
```

```python
# Fonction pour vérifier la similarité
def has_similar_name(name, lycees_liste):
    for element in lycees_liste:
        # Calculer le ratio de similarité
        similarity_ratio = fuzz.ratio(name, element)
        # Si le ratio est supérieur à une certaine valeur (par exemple, 80), considérez-le comme une correspondance
        if similarity_ratio > 77:
            return True
    return False
```

```python
"""Création du dataframe df_meilleurs_lycées_paris contenant seulement le nombre de lycées qu'on a fixé"""

# Appliquer la fonction pour créer une colonne de vérification de similarité
df_lycées_paris['similar_name'] = df_lycées_paris['patronyme_modif'].apply(lambda x: has_similar_name(x, liste_meilleurs_lycées))

# Filtrer les données en fonction de la colonne de similarité
df_meilleurs_lycées_paris = df_lycées_paris[df_lycées_paris['similar_name'] | df_lycées_paris['patronyme_modif'].isin(liste_meilleurs_lycées)]

# Supprimer la colonne temporaire 'similar_name'
df_meilleurs_lycées_paris = df_meilleurs_lycées_paris.drop(columns=['similar_name'])
```

```python
"""Fonction qui ajoute aux appartements le lycée le plus proche"""

def ajouter_lycees(dvf, df_lycées):
    
    #dvf : geodataframe contenant la localisation de l'appartement dans une colonne 'geometry' au format WSG84
    #df_lycées : geodataframe contenant la localisation des lycées dans une colonne 'geometry' au format WSG84

    #la fonction ajoute trois colonnes à dvf : closest_lycee, dist_min_lycee
    
    # Convertir la colonne 'geometry' en type Point
    dvf['geometry'] = dvf['geometry'].apply(Point)
    df_lycées['geometry'] = df_lycées['geometry'].apply(Point)
    
    # Définir la projection équidistante (EPSG:3395)
    dvf = dvf.to_crs(epsg=3395)
    df_lycées = df_lycées.to_crs(epsg=3395)
        
    # Fonction pour trouver la station de métro la plus proche
    def find_nearest_lycée(row):
        nearest_lycée = df_lycées.geometry == nearest_points(row['geometry'], df_lycées.unary_union)[1]
        return pd.Series({
            'closest_lycee': df_lycées.loc[nearest_lycée, 'patronyme'].values[0],
            'dist_min_lycee': row['geometry'].distance(df_lycées.loc[nearest_lycée, 'geometry'].values[0]) / 1000,  # Conversion en kilomètres
        })
    
    # Appliquer la fonction pour chaque appartement
    tqdm.pandas()  # Barre de progression pour les itérations
    dvf[['closest_lycee', 'dist_min_lycee']] = dvf.progress_apply(find_nearest_lycée, axis=1)

    # Reconvertir en WGS 84 (EPSG:4326)
    dvf_result = dvf.to_crs(epsg=4326)
    
    return dvf_result
```

```python
"""Chargement fichier dvf"""

#chemin du fichier dvf
path_folder = '1.2) Données_concat/'
name_file = 'dvf_concat_2021_to_2023.geojson'
path_file = path_folder + name_file

#lecture du fichier dvf
dvf = gpd.read_file(path_file)

#chemin export
outer_path = '3.1) Données_enrichies/'
```

```python
"""Fonction qui applique ajouter_lycees et exporte le fichier"""

def ajout_lycee_export(dvf, outer_path):

    #on applique la fonction ajouter_lycees
    dvf_enrichi = ajouter_lycees(dvf, df_meilleurs_lycées_paris)

    #nom du fichier de sortie
    name_file_without_extension = os.path.splitext(name_file)[0]
    outer_name = f'{name_file_without_extension}_avec_lycees.geojson'

    #on exporte
    dvf_enrichi.to_file(outer_path+outer_name, driver='GeoJSON')

    return dvf_enrichi
```

```python
"""Ajouter les lycées au fichier dvf"""

dvf_enrichi = ajout_lycee_export(dvf, outer_path)
```
